## Ejemplo 14.4 Una ecuación de salarios uülizando datos de panel* (wagepan.dta)

```python
year = pd.Categorical(dfej14_2.year)
dfej14_2= pd.DataFrame(dfej14_2.set_index(['nr','year']))
dfej14_2['year'] = year
```

```python
dfej14_2.head()
```

```python
FE=PanelOLS.from_formula('lwage ~ 1+educ+black+hisp+exper+expersq+married+union+year+EntityEffects',data=dfej14_2,drop_absorbed=True).fit()
```

```python
from linearmodels import PooledOLS
POLS = PooledOLS.from_formula('lwage ~ 1 + educ + black + hisp + exper + expersq + married + union + year', data=dfej14_2).fit()
```

```python
from linearmodels import RandomEffects
RE = RandomEffects.from_formula('lwage ~ 1 + educ + black + hisp + exper + expersq + married + union + year', data=dfej14_2).fit()

```

```python
from linearmodels.panel import compare
print(compare({'Pooled':POLS, 'RE':RE, 'FE':FE}))
```
