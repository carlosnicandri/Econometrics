```python
import numpy as np
import pandas as pd
import os
import statsmodels.api as sm
import statsmodels.formula.api as smf
#os.chdir('F:/CBL Tech/763392/Programacion/Python/libros/wool/Introducción/Capítulo 3')
```

## Ejemplo 3.1 Determinantes de la nota media en la unlversidad GPA (gpa1.dta)

```python
dfej3_1= pd.read_stata('gpa1.dta')
dfej3_1.head()
```

```python
dfej3_1.columns
```

```python
regdfej3_1 = smf.ols('colGPA ~ hsGPA + ACT', data = dfej3_1).fit() 
print(regdfej3_1.summary())
```

## Ejemplo 3.2 Ecuaclón del salario por hora (wage1.dta)

```python
dfej3_2= pd.read_stata('wage1.dta')
dfej3_2.head()
```

```python
regdfej3_2 = smf.ols('np.log(wage) ~ educ + exper + tenure', data = dfej3_2).fit() 
print(regdfej3_2.summary())
```
